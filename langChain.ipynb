```python
import tiktoken

def num_tokens_from_string(string: str, encoding_name: str) -> int:
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens
```

```python
%%time
massive = list()
for cycle in range(len(dataset_small)):
    messages = [
        SystemMessage(
            content="Тебе дан  заголовок новости и сама новость. Напиши кратко содержание в одно предложение"
        ),
        HumanMessage(
            #content="Си Цзиньпин изменил структуру вооруженных сил Китая, усилив кибернетическую составляющую. Эксперты увидели признаки ставки Пекина на информационное противостояние.В китайской армии будут расформированы силы стратегической поддержки и создано новое подразделение — силы информационной поддержки. Такой приказ отдал председатель КНР Си Цзиньпин, сообщило правительство.Силы стратегической поддержки появились девять лет назад, в 2015 году, для координации усилий в кибервойне и политическом противостоянии. Прежде подчиненные этой структуре аэрокосмические и кибернетические подразделения станут самостоятельными и будут действовать параллельно с силами информационной поддержки.Новое подразделение будет оказывать «ключевую поддержку в координации создания и использования киберинформационной системы», подчеркнул Си Цзиньпин.Bloomberg называет эти изменения крупнейшей реорганизацией китайской армии с 2015 года.Bloomberg отмечает, что реорганизация происходит на фоне конкуренции с США за глобальное влияние, а также обвинений со стороны западных стран в хакерской активности. Изменения направлены на улучшение готовности вооруженных сил к «информационным» условиям войны, оценивает их бывший старший научный сотрудник Военно-морской исследовательской академии китайской армии Цао Вэйдун.Предыдущий командующий силами стратегической поддержки Цзюй Цяньшэн перестал появляться на публике в середине прошлого года. В 2023-м китайские власти провели масштабные чистки среди высшего командования после обнаружения широко распространившейся коррупции в вооруженных силах.Всего с момента прихода Си Цзиньпина к власти в 2012-м сотни старших офицеров армии и тысячи высших партийных чиновников были арестованы, исчезли или «покончили с собой» — их довели до суицида или обставили убийство как суицид, утверждала Politico. "
            content = dataset_small['Title'][cycle] + " " + dataset_small['Article Text'][cycle]
        ),
    ]
    if num_tokens_from_string(messages[0].content + messages[1].content, "cl100k_base") <= 4096:
        res = chat.invoke(messages)
    else:
        res = "token limit exceeded"
#print(res.response_metadata) #метаданные
    massive.append(res)
```

```python
massive.count("token limit exceeded")
```

```python
massive[0].content
```

```python
indexes_token_limits = []
for i, x in enumerate(massive):
    if x == "token limit exceeded":
        indexes_token_limits.append(i)
```

```python
indexes_token_limits
```

```python
new_clear_dataset = dataset_small.copy()
for change_description in range(len(new_clear_dataset)):
    item = massive[change_description]
    if isinstance(item, str):
        new_clear_dataset['Description'][change_description] = item
    else:
        new_clear_dataset['Description'][change_description] = str(item.content)
new_clear_dataset
```

```python
 massive[change_description].content
```

```python
for drop_cycle in indexes_token_limits:
    new_clear_dataset.drop(drop_cycle,axis = 0, inplace = True)
new_clear_dataset
```
